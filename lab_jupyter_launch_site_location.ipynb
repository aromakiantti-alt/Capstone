Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

import folium
from folium.features import DivIcon

# --- Select launch site: CCAFS LC-40 ---
site_row = launch_sites_df[launch_sites_df['Launch Site'] == 'CCAFS LC-40'].iloc[0]
launch_site_lat = float(site_row['Lat'])
launch_site_lon = float(site_row['Long'])

# --- Your picked coastline coordinate from MousePosition (replace if needed) ---
coastline_lat = 28.56252
coastline_lon = -80.56753

# --- Compute distance (KM) using the provided function ---
distance_km = calculate_distance(launch_site_lat, launch_site_lon,
                                 coastline_lat, coastline_lon)

# --- Build/refresh a focused map so both points are visible ---
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Create and add a folium.Marker on your selected closest coastline point on the map
folium.Marker(
    [coastline_lat, coastline_lon],
    popup="Closest Coastline",
    icon=folium.Icon(color="green", icon="ok-sign")
).add_to(site_map)

# Also mark the launch site
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="Launch Site: CCAFS LC-40",
    icon=folium.Icon(color="blue", icon="info-sign")
).add_to(site_map)

# Draw the line between launch site and coastline
coordinates = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]
folium.PolyLine(locations=coordinates, color="orange", weight=3, opacity=0.9).add_to(site_map)

# Display the distance between coastline point and launch site using the icon property
midpoint = [(launch_site_lat + coastline_lat)/2, (launch_site_lon + coastline_lon)/2]
distance_marker = folium.Marker(
    midpoint,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' %
             "{:10.2f} KM".format(distance_km),
    )
)
distance_marker.add_to(site_map)

# Fit map to show both points clearly
site_map.fit_bounds(coordinates)

# In IBM Skills Network Labs: save and open the HTML file to view the interactive map
site_map.save("CCAFS_LC40_distance.html")
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python

site_row = launch_sites_df[launch_sites_df['Launch Site'] == 'CCAFS LC-40'].iloc[0]
launch_site_lat, launch_site_lon = float(site_row['Lat']), float(site_row['Long'])

city    = [28.61075, -80.80994]
railway = [28.57095, -80.80444]
highway = [28.562417, -80.570576]

m = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

MousePosition(position='topright', separator=' | ', num_digits=6, prefix='Lat/Lon:').add_to(m)

folium.CircleMarker([launch_site_lat, launch_site_lon], radius=5, color="blue",  fill=True, popup="CCAFS LC-40").add_to(m)
folium.CircleMarker(city,    radius=5, color="purple", fill=True, popup=f"City {city}").add_to(m)
folium.CircleMarker(railway, radius=5, color="gray",   fill=True, popup=f"Railway {railway}").add_to(m)
folium.CircleMarker(highway, radius=5, color="black",  fill=True, popup=f"Highway {highway}").add_to(m)

segments = [
    [[launch_site_lat, launch_site_lon], city],
    [[launch_site_lat, launch_site_lon], railway],
    [[launch_site_lat, launch_site_lon], highway],
]
colors = ["purple", "gray", "black"]

for seg, col in zip(segments, colors):
    folium.PolyLine(locations=seg, color=col, weight=3, opacity=0.9).add_to(m)

all_points = [pt for seg in segments for pt in seg]
m.fit_bounds(all_points)
dist_city    = calculate_distance(launch_site_lat, launch_site_lon, city[0],    city[1])
dist_railway = calculate_distance(launch_site_lat, launch_site_lon, railway[0], railway[1])
dist_highway = calculate_distance(launch_site_lat, launch_site_lon, highway[0], highway[1])

def midpoint(a, b):
    return [(a[0] + b[0]) / 2, (a[1] + b[1]) / 2]

pairs = [
    ([launch_site_lat, launch_site_lon], city,    dist_city),
    ([launch_site_lat, launch_site_lon], railway, dist_railway),
    ([launch_site_lat, launch_site_lon], highway, dist_highway),
]

for start, end, dist_km in pairs:
    folium.Marker(
        location=midpoint(start, end),
        icon=DivIcon(
            icon_size=(120, 24),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{dist_km:.2f} KM</b></div>',
        ),
    ).add_to(m)


m.save("ccafs_lc40_three_lines_verified_endpoints.html")
m
```
